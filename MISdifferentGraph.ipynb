```python
# util_usage.ipynb

# Import the utilities
from util import (
    generate_random_graph, visualize_graph, mis, 
    global_maximum_independent_set, generate_complete_graph, 
    generate_path_graph, generate_cycle_graph, generate_star_graph, 
    generate_tree_graph, generate_grid_graph
)

# Define the number of nodes
num_nodes = 10

# Generate different types of graphs
graph_types = {
    "Complete Graph": generate_complete_graph(num_nodes),
    "Path Graph": generate_path_graph(num_nodes),
    "Cycle Graph": generate_cycle_graph(num_nodes),
    "Star Graph": generate_star_graph(num_nodes),
    "Tree Graph": generate_tree_graph(num_nodes),
    "Grid Graph": generate_grid_graph(3, 3)  # For grid graph, ensure the product of dimensions equals num_nodes
}

# User choice for visualization
visualize = False  # Set this to True if you want to visualize the graphs

# Compare maximum independent sets for different graph structures
mis_results = {}
for graph_type, G in graph_types.items():
    heuristic_mis = mis(G, visualize=visualize)
    global_mis = global_maximum_independent_set(G)
    mis_results[graph_type] = {
        "Heuristic MIS": heuristic_mis,
        "Global MIS": global_mis
    }

# Print the results
for graph_type, result in mis_results.items():
    print(f"{graph_type}:")
    print(f"  Heuristic MIS: {result['Heuristic MIS']}")
    print(f"  Global MIS: {result['Global MIS']}")

# Visualize the graphs if the user chose to visualize
if visualize:
    for graph_type, G in graph_types.items():
        heuristic_mis = mis_results[graph_type]["Heuristic MIS"]
        global_mis = mis_results[graph_type]["Global MIS"]
        visualize_graph(G, f"{graph_type} with Heuristic Maximum Independent Set Highlighted", heuristic_mis)
        visualize_graph(G, f"{graph_type} with Global Maximum Independent Set Highlighted", global_mis)
```
