```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_graph(n):
    # Create a new graph
    G = nx.Graph()

    # Add nodes
    G.add_nodes_from(range(1, n+1))

    # Add edges
    for i in range(1, n-1):
        G.add_edge(i, i+1)
        G.add_edge(i, i+2)

    # Add the last edge to complete the pattern
    if n > 2:
        G.add_edge(n-1, n)

    # Set up the plot
    plt.figure(figsize=(12, 8))
    
    # Draw the graph
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', 
            node_size=500, font_size=16, font_weight='bold')
    
    # Add edge labels
    edge_labels = {(u, v): f'{u}-{v}' for (u, v) in G.edges()}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

    # Add a title
    plt.title(f"Graph with {n} nodes", fontsize=16)
    
    # Show the plot
    plt.axis('off')
    plt.tight_layout()
    plt.show()

# Example usage
draw_graph(15)
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

def draw_graph(n):
    # Create an adjacency matrix
    adj_matrix = np.zeros((n, n), dtype=int)
    
    # Populate the adjacency matrix with edges
    for i in range(n-1):
        adj_matrix[i, i+1] = 1  # Edge from i to i+1
        if i+2 < n:
            adj_matrix[i, i+2] = 1  # Edge from i to i+2

    # Create a NetworkX graph from the adjacency matrix
    G = nx.from_numpy_array(adj_matrix)
    
    # Set up the plot
    plt.figure(figsize=(12, 8))
    
    # Draw the graph
    pos = nx.spring_layout(G)  # Spring layout positions nodes based on the structure of the graph
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=16, font_weight='bold')
    
    # Add edge labels
    edge_labels = {(u, v): f'{u}-{v}' for u, v in G.edges()}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

    # Add a title
    plt.title(f"Graph with {n} nodes", fontsize=16)
    
    # Show the plot
    plt.axis('off')
    plt.tight_layout()
    plt.show()

    return G , adj_matrix

# Example usage
G , adj_matrix = draw_graph(10)
```
